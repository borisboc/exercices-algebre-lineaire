## Question 1

Pour quelles valeurs de $\alpha$ les vecteurs

$$
v_{1} = (1,-1,0,2), v_{2} = (1,0,1,2), v_{3} = (1,3,5,7), v_{4} = (0,2,3,\alpha) 
$$

forment-ils une base de $\mathbb{R}^{4}$ ? 

```python
from sympy import Matrix
from sympy import symbols
```

Regroupons tous les vecteurs sous la forme d'une matrice "ligne" $V_{s}$

```python
v1 = Matrix(1, 4, [1, -1, 0, 2])
v2 = Matrix(1, 4, [1, 0, 1, 2])
v3 = Matrix(1, 4, [1, 3, 5, 7])
a = symbols('alpha')
v4 = Matrix(1, 4, [0, 2, 3, a])
Vs = Matrix.vstack(v1, v2, v3, v4)
Vs
```

Appliquons la technique du pivot (ou élimination de Gauss) pour échelonner la matrice $V_{s}$.


Pour échelonner la ligne correspondant à $V_{2}$, nous allons poser $$V_{2}^{\prime} = V_{2} - V_{1}$$

```python
Vs_pivot = Vs.copy()
Vs_pivot[1,:] = v2-v1
Vs_pivot
```

Pour échelonner la ligne correspondant à $V_{3}$, nous allons poser 

$$
\begin{align}
V_{3}^{\prime} &= V_{3} - V_{1} - 4V_{2}^{\prime} \\
\iff V_{3}^{\prime} &= V_{3} - V_{1} - 4(V_{2} - V{1}) \\
\iff V_{3}^{\prime} &= V_{3} + 3V_{1} - 4V_{2} \\
\end{align}
$$


```python
Vs_pivot[2,:] = v3-v1-4*Vs_pivot[1,:]
Vs_pivot
```

Pour échelonner la ligne correspondant à $V_{4}$, nous allons poser
$$
\begin{align}
V_{4}^{\prime} &= V_{4} - 2V_{2}^{\prime} - V_{3}^{\prime} \\
\iff V_{4}^{\prime} &= V_{4} - 2(V_{2} - V_{1}) - (V_{3} + 3V_{1} - 4V_{2}) \\
\iff V_{4}^{\prime} &= V_{4} - V_{1} + 2V_{2}  -V_{3} \\
\end{align}
$$

```python
Vs_pivot[3,:] = v4-2*Vs_pivot[1,:]-Vs_pivot[2,:]
Vs_pivot
```

Vérifions que la démarche est correct en utilisant les fonctionnalités de SymPy

```python
Vs.echelon_form()
```

```python
Vs_pivot == Vs.echelon_form()
```

Notre démarche retourne le même résultat que l'implémentation de SymPy.

La matrice est maintenant échelonnée. On peut voir que la dimension de l'image de $V_{s}$ est bien 4 pour tout réel $\alpha$  privé de la valeur 5. Ce qu'on peut écrire 

$$
\forall \alpha \in \mathbb{R} \backslash \{5\}
$$

Dans un soucis de factorisation (puisque ces opérations vont être répétées à la question 3), regroupons ces opérations d'échelonnnage dans une fonction python.

```python
def echelonnage_question_1(mat_in : Matrix) -> Matrix:
    mat_ech = mat_in.copy()
    mat_ech[1,:] = mat_in[1,:]-mat_in[0,:]
    mat_ech[2,:] = mat_in[2,:]-mat_in[0,:]-4*mat_ech[1,:]
    mat_ech[3,:] = mat_in[3,:]-2*mat_ech[1,:]-mat_ech[2,:]
    return mat_ech
```

## Question 3

Soit $V = (-2, k, 1, 3)$. Pour quelles valeurs de $k$, $V \in Vect\{ V_{1}, V_{2}, V_{3}, V_{4}\}$ ? Déterminer, dans ce cas, les composantes de $V$ dans une base de $Vect\{ V_{1}, V_{2}, V_{3}, V_{4}\}$.

Posons le problème mathématique équivalent à résoudre : 

 $$
 \exist ? k \in \R | [ \exist  (\lambda_{1}, \lambda_{2}, \lambda_{3}, \lambda_{4}) \in \R^{4} | V = (-2, k, 1, 3) = \lambda_{1}V_{1} + \lambda_{2}V_{2} + \lambda_{3}V_{3} + \lambda_{4}V_{4} ]
 $$ 


On peut écrire la seconde partie de la proposition sous la forme matricielle suivante : 

$$
(\lambda_{1}, \lambda_{2}, \lambda_{3}, \lambda_{4}) * V_{s} = (-2, k, 1, 3)
$$

Avec la même matrice "ligne" $V_{s}$ qu'au premier exercice.

Nous allons re-utiliser la technique du pivot (ou élimination de Gauss) pour échelonner la matrice $V_{s}$, sauf que cette fois nous devons également appliquer les différentes transformations sur le second terme non nul $(-2, k, 1, 3)$ .


```python
k = symbols("k")
V = Matrix(1, 4, [-2, k, 1, 3])
V
```

Nous allons utiliser la fonction python implémentée et validée à la question 1.

```python
V_pivot = echelonnage_question_1(V.transpose())
V_pivot
```

En utilisant la version échelonnée de $V_{s}$ et de $V$, nous avons le système d'équations équivalentes : 

$$

(\lambda_{1}, \lambda_{2}, \lambda_{3}, \lambda_{4}) * 


\begin{pmatrix}

1 & -1 & 0 & 2 \\
0 & 1 & 1 & 0 \\
0 & 0 & 1 & 5 \\
0 & 0 & 0 & \alpha -5 \\
\end {pmatrix}


 = 


\begin{pmatrix}
-2 \\
k + 2 \\
-4k - 5 \\
2k + 4
\end {pmatrix}



$$

En substituant de "bas en haut", c'est à dire de $\lambda_{4}$ à $\lambda_{1}$, nous allons obtenir successivement : 

$$
\begin{align}
\lambda_{4} &= \frac{2k+4}{\alpha-5} \\
\lambda_{3} &= -4k - 5 - 5 \lambda_{4} \\
\lambda_{2} &= k + 2 - \lambda_{3} = 5k + 7 + 5 \lambda_{4} \\
\lambda_{1} &= -2 + \lambda_{2} - 2 \lambda_{4} = 5k + 5 + 3 \lambda_{4}
\end{align}
$$
