## Question 1

Pour quelles valeurs de $\alpha$ les vecteurs

$$
v_{1} = (1,-1,0,2), v_{2} = (1,0,1,2), v_{3} = (1,3,5,7), v_{4} = (0,2,3,\alpha) 
$$

forment-ils une base de $\mathbb{R}^{4}$ ? 

```python
from sympy import Matrix
from sympy import symbols
```

Regroupons tous les vecteurs sous la forme d'une matrice "colonne" $V_{s}$

```python
v1 = Matrix(4, 1, [1, -1, 0, 2])
v2 = Matrix(4, 1, [1, 0, 1, 2])
v3 = Matrix(4, 1, [1, 3, 5, 7])
a = symbols('alpha')
v4 = Matrix(4, 1, [0, 2, 3, a])
Vs = Matrix.hstack(v1, v2, v3, v4)
Vs
```

Pourquoi une matrice où les vecteurs sont placés en "colonne" plutôt qu'en "ligne", comme le suggère le livre, au chapitre 2 ? Premièrement, en remarque, le livre, dit qu'on peut aussi travailler avec une matrices "colonne". Le livre à choisi de faire en ligne, puisque faire des opérations sur lignes d'un système lors de la méthode du pivot, revient à faire des opérations sur les vecteurs (puisqu'ils sont disposés en ligne). Néanmoins, cette disposition en ligne peut être problématique car elle ne représente pas le même système d'équation, avec le second terme (le "right hand side") en colonne. Je tente d'expliquer ceci ci dessous.

Disons que nous voulons travailler sur problème suivant : 

$$
w_{lhs} V_{1} + x_{lhs} V_{2} + y_{lhs} V_{3} + z_{lhs} V_{4} = (w_{rhs}, x_{rhs}, y_{rhs}, z_{rhs})
$$

Nous pouvons le représenter sous la forme du système d'équations ci dessous : 

$$ \begin{cases}
w_{lhs} &+ x_{lhs} &+ y_{lhs} & &= w_{rhs} \\
-w_{lhs} & & + 3y_{lhs} & +2z_{lhs} &= x_{rhs} \\
& x_{lhs} &+ 5y_{lhs} &+3z_{lhs} &= y_{rhs} \\
2w_{lhs} &+ 2x_{lhs} &+ 7y_{lhs} &+\alpha &= z_{rhs} \\
\end{cases} $$ 

Si on concatène nos vecteurs en ligne dans une matrice "ligne", nous devons multiplier "à droite" le vecteur dit "left hand side" (lhs) aussi sous forme de vecteur ligne, pour avoir un second terme (dit "right hand side", abrégé "rhs") lui aussi en ligne. Comme le montre la formule ci dessous.


$$
(w_{lhs}, x_{lhs}, y_{lhs}, z_{lhs}) * 

\begin{pmatrix}
1 & -1 & 0 & 2 \\
1 & 0 & 1 & 2 \\
1 & 3 & 5 & 7 \\
0 & 2 & 3 & \alpha 
\end{pmatrix}

= (w_{rhs}, x_{rhs}, y_{rhs}, z_{rhs})
$$

Or, appliquer la méthode du pivot sur cette matrice ligne, va certes permettre de répondre à la première question (comme nous le montrerons plus bas), mais les différentes opérations successives appliquées pour échelonner la matrice ne sont pas applicables sur le second terme.

Ainsi, pour "mieux" représenter le système d'équations avec un second terme (right hand side) en colonne, et un produit matricielle "à gauche", il semble préférable d'utiliser une matrice "colonne" comme montré ci dessous :


$$
\begin{pmatrix}
1 & 1 & 1 & 0 \\
-1 & 0 & 3 & 2 \\
0 & 1 & 5 & 3 \\
2 & 2 & 7 & \alpha 
\end{pmatrix}
*
\begin{pmatrix}
w_{lhs} \\
x_{lhs} \\
y_{lhs} \\
z_{lhs} 
\end{pmatrix}
= 
\begin{pmatrix}
w_{rhs} \\
x_{rhs} \\
y_{rhs} \\
z_{rhs} 
\end{pmatrix}
$$

Formulation dont les opérations pour échelonner seront également ré-utilisables pour la question 3, plus bas.

Appliquons la technique du pivot (ou élimination de Gauss) pour échelonner la matrice $V_{s}$.
Notons

$$
\begin{align}
L_{1} &= (1, 1, 1, 0) \\
L_{2} &= (-1, 0, 3, 2) \\
L_{3} &= (0, 1, 5, 3) \\
L_{4} &= (2, 2, 7, \alpha) \\
\end{align}
$$

Les lignes de la matrice "colonne" $V_{s}$

Pour échelonner la ligne correspondant à $L_{2}$, nous allons poser 
$$
L_{2}^{\prime} = L_{2} + L_{1}
$$

```python
Vs_pivot = Vs.copy()
Vs_pivot[1,:] = Vs[1, :] + Vs[0, :]
Vs_pivot
```

Pour échelonner la ligne correspondant à $L_{3}$, nous allons poser 
$$
L_{3}^{\prime} = L_{3} - L_{2}^{\prime}
$$


```python
Vs_pivot[2,:] = Vs[2, :] - Vs_pivot[1,:] 
Vs_pivot
```

Pour échelonner la ligne correspondant à $L_{4}$, nous allons poser
$$
\begin{align}
L_{4}^{\prime} &= L_{4} - 2L_{1} - 5L_{3}
\end{align}
$$

```python
Vs_pivot[3,:] = Vs[3, :] -2*Vs_pivot[0,:] - 5*Vs_pivot[2,:]
Vs_pivot
```

Vérifions que la démarche est correct en utilisant les fonctionnalités de SymPy

```python
Vs.echelon_form()
```

```python
Vs_pivot == Vs.echelon_form()
```

Notre démarche retourne le même résultat que l'implémentation de SymPy.

La matrice est maintenant échelonnée. On peut voir que la dimension de l'image de $V_{s}$ est bien 4 pour tout réel $\alpha$  privé de la valeur 5. Ce qu'on peut écrire 

$$
\forall \alpha \in \mathbb{R} \backslash \{5\}
$$

Dans un soucis de factorisation (puisque ces opérations vont être répétées à la question 3), regroupons ces opérations d'échelonnnage dans une fonction python.

```python
def echelonnage_question_1(mat_in : Matrix) -> Matrix:
    mat_ech = mat_in.copy()
    mat_ech[1,:] = mat_in[1,:] + mat_in[0,:]
    mat_ech[2,:] = mat_in[2,:]- mat_ech[1,:]
    mat_ech[3,:] = mat_in[3,:] -2*mat_ech[0,:] - 5*mat_ech[2,:]
    return mat_ech
```

Vérification de l'implémentation : 

```python
Vs_pivot2 = echelonnage_question_1(Vs)
Vs_pivot2
```

```python
Vs_pivot2 == Vs_pivot
```

Implémentation vérifiée
